```python
# convert maize to geodataframe
maize_hist = gpd.GeoDataFrame(maize_hist, geometry=gpd.points_from_xy(maize_hist.long_maize_hist, maize_hist.lat_maize_hist))
maize_370 = gpd.GeoDataFrame(maize_370, geometry=gpd.points_from_xy(maize_370.long_maize_370, maize_370.lat_maize_370))
maize_126 = gpd.GeoDataFrame(maize_126, geometry=gpd.points_from_xy(maize_126.long_maize_126, maize_126.lat_maize_126))
```
